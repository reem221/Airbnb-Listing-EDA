```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df=pd.read_csv('listings.csv')
df.head()
```

```python
# Display summary information
df.info()
df.describe()
```

```python
# Clean the price column by removing dollar signs and commas, then convert to numeric
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Now you can safely handle missing values and outliers
df = df.dropna(subset=['price', 'availability_365', 'latitude', 'longitude'])

# Remove outliers by keeping only 95% of the data
df = df[df['price'] < df['price'].quantile(0.95)]
```

```python
# Import necessary libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
# Price Analysis
# Set up the figure for plotting with a specified size
plt.figure(figsize=(8, 5))
# Create a distribution plot of the 'price' column from the DataFrame 'df'
# with a Kernel Density Estimate (KDE) to show the distribution shape
sns.histplot(df['price'], bins=20, kde=True, color='lightblue')
# Add a title and labels to the plot
plt.title('Price Distribution of Airbnb Listings', fontsize=16)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
# Save the plot as an image file in the specified directory
plt.savefig('E:\\reem\\electro pi\\Exploratory data\\Distribution of Prices.png')
# Display the plot on the screen
plt.grid(True)
plt.tight_layout()
plt.show()
print('Price Distribution: "The price distribution shows that most Airbnb listings fall within the $50-$200 range, with a few outliers in the higher price range. This suggests that the majority of listings cater to budget and mid-range travelers."')
```

```python
from scipy import stats

# Get price statistics
mean_price = df['price'].mean()
median_price = df['price'].median()
std_price = df['price'].std()

print(f"Mean price: {mean_price}")
print(f"Median price: {median_price}")
print(f"Standard deviation of price: {std_price}")

# Perform a T-test for two property types (e.g., apartments vs. houses)
apartment_prices = df[df['property_type'] == 'Apartment']['price']
house_prices = df[df['property_type'] == 'House']['price']
t_stat, p_value = stats.ttest_ind(apartment_prices, house_prices, equal_var=False)

print(f"T-test between apartments and houses:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# availability Analysis
# Set up the figure for plotting with a specified size
plt.figure(figsize=(8,5))
# Create a distribution plot of the 'availability_365' column from the DataFrame 'df'
# with a Kernel Density Estimate (KDE) to show the distribution shape
sns.histplot(df['availability_365'], bins=20, kde=True, color='lightgreen')
# Add a title and labels to the plot
plt.title('Availability of Airbnb Listings (Days per Year)', fontsize=16)
plt.xlabel('Availability (Days)', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
# Save the plot as an image file in the specified directory
plt.savefig('E:\\reem\\electro pi\\Exploratory data\\availability_distribution.png')
# Display the plot on the screen
plt.grid(True)
plt.tight_layout()
plt.show()
print("'Analysis for this graphic:' The graphic shows the distribution of the number of days Airbnb listings are available during the year. Several listings are available continuously throughout the year, while others are available for shorter periods.")
```

```python
from scipy.stats import linregress
# Correlation between price and availability_365
corr = df['availability_365'].corr(df['price'])

print(f"Correlation between price and availability: {corr}")

# Perform a linear regression
slope, intercept, r_value, p_value, std_err = linregress(df['availability_365'], df['price'])

print(f"Linear Regression between price and availability:")
print(f"Slope: {slope}, Intercept: {intercept}")
print(f"R-squared: {r_value**2:.2f}")
print(f"P-value: {p_value}")
```

```python
# Location Analysis (using latitude and longitude)
# Set up the figure for plotting with a specified size
plt.figure(figsize=(8,5))
# Create a scatter plot showing the relationship between 'longitude' and 'latitude'
# The color of the points is determined by the 'price' column, using the 'viridis' color palette
sns.scatterplot(x='longitude', y='latitude', hue='price', data=df, palette='coolwarm', s=50)
# Add a title and labels to the plot
plt.title('Airbnb Listings by Location and Price', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
# Save the plot as an image file in the specified directory
plt.savefig('E:\\reem\\electro pi\\Exploratory data\\Price Distribution by Location.png')
# Display the plot on the screen
plt.grid(True)
plt.tight_layout()
plt.show()
print ('Analysis for this graphic: "This graph shows the distribution of Airbnb listing prices. Most listings fall within a certain price range, with a few being significantly more expensive."')
```

```python
# Property Type Analysis
# Set up the figure for plotting with a specified size
plt.figure(figsize=(10, 6))
# Create a count plot for the 'property_type' column from the DataFrame 'df'
# The order of the bars is determined by the frequency of each property type
sns.countplot(y='property_type', data=df, order=df['property_type'].value_counts().index)
# Add a title and labels to the plot
plt.title('Distribution of Property Types in Airbnb Listings', fontsize=16)
plt.xlabel('Number of Listings', fontsize=12)
plt.ylabel('Property Type', fontsize=12)
# Add gridlines for better readability
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
# Save the plot as an image file in the specified directory
plt.savefig('E:\\reem\\electro pi\\Exploratory data\\property_type_distribution.png')
# Display the plot
plt.tight_layout()
plt.show()

# Insight: Property Type Distribution
print("Insight: The count plot shows the distribution of different property types listed on Airbnb. It reveals that certain property types, like apartments and houses, dominate the listings, while less common types like condos or lofts have fewer listings.")
```

```python
from scipy.stats import chi2_contingency

# Perform a Chi-Square test on the property type counts
property_type_counts = df['property_type'].value_counts()
chi2, p, dof, expected = chi2_contingency([property_type_counts])

print(f"Chi-Square Test for Property Types:")
print(f"Chi2 statistic: {chi2}")
print(f"P-value: {p}")

if p < 0.05:
    print("There is a significant difference in the distribution of property types.")
else:
    print("The distribution of property types is not significantly different.")
```

```python
# Select all columns from the DataFrame 'df' that have numeric data types
# and store them in a new DataFrame 'data'
data = df.select_dtypes(include='number')

# Correlation Analysis
# Set up the figure for plotting with a large size
plt.figure(figsize=(12, 10))
# Create a heatmap of the correlation matrix for the numeric columns in 'data'
# The 'annot=True' parameter adds numerical annotations to each cell in the heatmap
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
# Add a title to the plot
plt.title('Correlation Matrix of Numeric Features', fontsize=16)
# Save the plot as an image file in the specified directory
plt.savefig('E:\\reem\\electro pi\\Exploratory data\\correlation_matrix.png')
# Display the plot
plt.tight_layout()
plt.show()

# Insight: Correlation Analysis
print("Insight: The heatmap shows the correlation between different numeric variables in the dataset. Darker colors represent stronger correlations. For example, price and availability do not seem highly correlated, but other variables may exhibit stronger relationships.")
```
